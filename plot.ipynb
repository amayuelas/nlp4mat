```python
import matplotlib.pyplot as plt
```

```python

# Extracting property values
properties = {
    "Temperature (K)": 2.0,
    "Pressure (GPa)": 1.0,
    "Magnetic Field (T)": 9.0
}

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(properties.keys(), properties.values(), color=['blue', 'green', 'red'])
plt.xlabel('Properties')
plt.ylabel('Values')
plt.title('Measured Properties of Sr2MgReO6')
plt.yscale('log')  # Using log scale for better visualization of varying magnitudes
plt.show()
```
